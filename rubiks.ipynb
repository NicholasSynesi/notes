Any geomtric deep learning problem should respect the symmetries of the data it consumes. While this line is quite poetic, what does it actually mean? To illustrate the power of what I have said, I present the 2x2x2 Rubik's cube.

The 2x2x2 Rubik's cube (I will now just refer to this as the 'cube') consists of 8 corner cubies. Any configuration can be specified by:

- Permutation. Which physical corner occupies which of the 8 corner slots. 

- Orientation. Each corner, relative to a fixed reference, is rotated by one of three amounts.

The 'group theory' section this time involves being far more subtle. Let's go through this one section at a time.

### Permutation Group

Hopefully this is beginning to feel like a familiar one now, it's the one that asks what happens when I start shuffling the cubies around? How many ways can this be done? It's as easy as seeing there is 8! possible combinations of each of these pieces, easy. This represents the S8 group.

### Orientation Group

Now this is trickier, because even when a cubie is sat in one of the 8 positions, it's actually got 3 different angles it can sit at, so each of the 3 colours on each corner can be in each position. Let's say each cubie can be in the {0, 1, 2} state, then this represents a Z3 group. But this is just one cubie, each of them are indepedent of each other. So we actually need 8 copies of Z3, which is just (Z3)^8. 

So, the entire problem lives in S8 & (Z3)^8.

### The reachable space

It turns out however, that this is actually a somewhat naive approach, as there are some constraints on our system. Namely:

- Twist constaint. The sum of all eight corner twists is zero, i.e if you twist every piece, you're back where you started. This means only 3^7 out of the 3^8 twist vectors are reachable.

- Gauge freedom. The entire cube can be rotated in space without changing anything about how 'solved' it is. This means one corner can be fixed as a reference frame. Fixing a corner leaves 7 movable corners.

So, the reachable space has size 7! x 3^6 = 3,674,160 states. That's actually quite a lot, and this is just for a small cube!

The diameter of this space under the quarter-turn metric (God's number) is actually only 11 (An angel number?). So despite all those many states, any cube is actually at MOST, 11 moves from solved. 

## There are actually two symmetry groups

I know that may seem sudden, but hear me out! We have to separate them as they act differently and a model may exploit either. 

#### Spatial rotation group of the cube

A cube has 24 orientation-preserving symmetries. Relabeling the cube's frame by any of these 24 rotations does not change any intrinsic property such as 'solved-ness'. This is an invariant symmetry.

####  The move group

These are the generators of legal transformations. Equivariance under this group is more powerful, but hard to realise architecturally. For this reason, we'll stuck to the spatial rotation.

## Orbits!

As usual, we calculate the number of orbits using Burnside's lemma. Looking at the corners again, we can see that any corner can be carried to another by some sort of spatial rotation. So there is one orbit here. This means that, for the ansatz we create, all entanglers will share just one parameter, which is want we want! This may however, prove slightly *too* rigid, alas we will see.

## The circuit

Each cubie can be in one of three states, {0, 1, 2}, so we apply the initial rotation gate at an angle of 2pi/3 (Eagle eyed fans will see this is the same as the Connect 4 game, and it is! Each cell there was in one of the states {-1, 0, 1}). 

Following from here is somewhat tricky, as the permutation state is discrete and combinatorial. This means amplitude based quantum encodings don't naturally represent "which slot holds which corner". We can just take the 7 corners and 7 slots as 49 qubits but this is very expensive (all jokes aside, my ThinkPad literally could not do this at all. It would receive the problem and then just pop out of existence, sad!)

Alternatively, let a corner's identity be encoded by which physical qubits carry its twist register, so the permuting corners permute qubit registers. Bit of a tongue-twister there, but this makes the spatial symmetry act naturally, even if it leads to a messy gate layout. 

From what I can see, this seems to be an unsolved engineering problem. If someone can get on this so I can solve the all important Rubik's cube, that'd be great.

How about the equivariant approach I've been taking. Only entangle cimponents that share meaningful structure, and tie their weights across symmetry orbits. This would lead to the following:

- Encoding layer, the 2pi/3 twist
- Equivariant single-corner layer. Parameterised RX, RY on each corner register, with weights shared across the corner orbits. 
- Equivariant entangling layer. CRY gates between corners sharing a face, with weights tied across the orbits of face-adjacent corner pairs under the chosen group
- Invariant readout. Symmetry-invariant measurement of the summed Pauli-Z over all qubits

So honestly? Nothing we haven't seen before here. Few initial rotations, couple entanglements, a readout or two. Bosh, as they say.

## It's never that easy...

The conceptual result  here is unfortunately negative and instructive. Maximal symmetrisation imposing the full 24-element rotation group collapses the corners into a single orbit, yielding a model that has too few paramters to represent non trivial functions of the cube state. I overcooked it! Symmetry reduction is a powerful tool, but situational.

However, all is not lost. We must see the results ourself. We will proceed! 

On a serious note, this is actually a toy model of a Z3 gauge system. The 2pi/3 factor is literally encodin a Z3 charge onto a qubit, which is a key result needed in lattice-gauge QML. Cool.
